```python
from netCDF4 import Dataset
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
# nc파일 read
data = Dataset(r'/Users/songchangseok/big/data2/OSTIA_data.nc', 'r')
# nc파일에 있는 각 변수의 데이터들을 정의

# 잘 변환 됐는지 확인하기 위해서 for문 이용
# for i in days:
#    start = ref + dt.timedelta(i) # 기준점으로 부터 데이터만큼 지난 날짜와 시간으로 변환
#    print(start)
# --------------------------------
lats = data.variables['lat'][:]
lons = data.variables['lon'][:]
time = data.variables['time'][:]
sst = data.variables['analysed_sst'][:]

mp = Basemap(projection= 'merc',
             llcrnrlon=125.825,
             llcrnrlat=33.925,
             urcrnrlon=133.225,
             urcrnrlat=42.675,
             resolution='i')
lon, lat = np.meshgrid(lons, lats)
x, y = mp(lon, lat)
days = np.arange(0,365)
for i in days:
    c_scheme = mp.pcolor(x,y,np.squeeze(sst[i,:,:]),cmap = 'jet')
    mp.drawcoastlines()
    mp.drawstates()
    mp.drawcountries()
    day = i+1
    plt.title('Average temparature on ' + str(day))
    plt.savefig(str(day)+'.jpg')
    plt.clf()
    plt.show()
```

```python
from mpl_toolkits.basemap import Basemap, shiftgrid, cm
import numpy as np
import matplotlib.pyplot as plt
from netCDF4 import Dataset

# The netCDF file is from ftp://ftp.cdc.noaa.gov/Datasets/noaa.ersst/sst.mnmean.nc
d = Dataset(r'/Users/songchangseok/big/data2/OSTIA_data.nc', 'r')
sst = d.variables["analysed_sst"][:][0, :, :]  # Should take the first time slice.
lon = d.variables["lon"][:]
lat = d.variables["lat"][:]
sst, lon = shiftgrid(185, sst, lon, start = False)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# For north Polar Stereographic projection
#m = Basemap(projection='npstere',boundinglat=10,lon_0=270,resolution='l')
# For classic Mercator... in case you hate geography.
m = Basemap(projection='merc', llcrnrlat=-80, urcrnrlat=80,
            llcrnrlon=-180, urcrnrlon=180, lat_ts=20, resolution='c')
# 555660 is because this is a 5.0 x 5.0 degree grid; 111132 meters are in about 111 km appart. This is fairly rough.
nx = int((m.xmax - m.xmin)/555660); ny = int((m.ymax - m.ymin)/555660)
sst = m.transform_scalar(sst, lon, lat, nx, ny)
im = m.imshow(sst, interpolation = "none")

m.drawcoastlines()
parallels = np.arange(-90, 90, 30)
meridians = np.arange(-180, 180, 60)
m.drawparallels(parallels, labels = [1, 0, 0, 1])
m.drawmeridians(meridians, labels = [1, 0, 0, 1])

cb = m.colorbar(im, "right", size = "5%", pad = "2%")
ax.set_title("Kaplan ERSST January 1895")
plt.show()
 
```
